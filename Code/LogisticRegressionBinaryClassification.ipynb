```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("diabetes.csv")
df.head(10)
```

```python
%matplotlib inline
plt.scatter(df[["DiabetesPedigreeFunction"]],df[["Outcome"]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']],df.Outcome,test_size=0.1)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict_proba(x_test)
```
